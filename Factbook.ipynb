```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

conn = sqlite3.connect("factbook.db")
cursor = conn.cursor()
```

```python
q = "SELECT * FROM sqlite_master WHERE type='table';"
cursor.execute(q).fetchall()
```

```python
df = pd.read_sql_query(q, conn)
df
```

```python
q1 = "select * from facts"
df1 = pd.read_sql_query(q1, conn)
df1
```

```python
q2 = "select * from facts"
df = pd.read_sql_query(q2,conn)
df
```

```python
q3 = "select MIN(population) lowest_population,MAX(population) highest_population ,MIN(population_growth) lowest_pop_growth, MAX(population_growth) highest_pop_growth from facts;"
df3 = pd.read_sql_query(q3,conn)
df3
```

```python
q4 = "Select * from facts where population = 0"
df4 = pd.read_sql_query(q4,conn)
df4
```

```python
q5 = "Select * from facts where population = 7256490011"
df5 = pd.read_sql_query(q5,conn)
df5
```

```python
q7 = "select population, population_growth, birth_rate, death_rate from facts where population != (select MAX(population) from facts) OR population != (select MIN(population) from facts)"
df7 = pd.read_sql_query(q7,conn)
df7
```
